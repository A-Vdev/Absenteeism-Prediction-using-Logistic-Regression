### Extract The Month Value

```python
df_reason_copy['Date'][0]
```

```python
df_reason_copy['Date'][0].month
```

```python
len(df_reason_copy['Date'])
```

```python
list_month=[]
```

```python
for i in range(len(df_reason_copy['Date'])):
    list_month.append(df_reason_copy['Date'][i].month)
          
len(list_month)
               
```

```python
df_reason_copy['Month Value']=list_month
```

```python
df_reason_copy.columns.values
```

```python
# Monday-Suday=0-6
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_copy["Weekday"]=df_reason_copy['Date'].apply(date_to_weekday)
```
